```python
def generate_relative_marks(num_students):
    marks = []
    for _ in range(num_students):
        student_marks = {
            "cat1": min(max(int(random.gauss(32, 15)), 0), 50),
            "cat2": min(max(int(random.gauss(32, 15)), 0), 50),
            "quiz1": min(max(int(random.gauss(6.5, 3)), 0), 10),
            "quiz2": min(max(int(random.gauss(6.5, 3)), 0), 10),
            "da": min(max(int(random.gauss(9, 1)), 0), 10),
            "fat": min(max(int(random.gauss(65, 25)), 0), 100),
            "total_marks": 0,
            "grade": 0
        }
        marks.append(student_marks)
    return marks

def generate_absolute_marks(num_students):
    marks = []
    for _ in range(num_students):
        student_marks = {
            "ia1": min(max(int(random.gauss(17, 3)), 0), 20),
            "mt": min(max(int(random.gauss(17, 3)), 0), 20),
            "ia2": min(max(int(random.gauss(17, 3)), 0), 20),
            "fat": min(max(int(random.gauss(32, 5)), 0), 40),
            "total_marks": 0,
            "grade": 0
        }
        marks.append(student_marks)
    return marks
```

```python
def fetch_student_data(semester_name, subject):
    query = {
        "semester_data": {
            "$elemMatch": {
            "semester_name": semester_name,
            "subject_list": subject
            }
        }
    }
    projection = {"register_number": 1, "_id": 0}
    results = students_collection.find(query, projection)
    register_numbers = [student["register_number"] for student in results]
    return register_numbers
```

```python
def format_class_data(class_number, subject_id, student_data, marks_data, subject_type):
    class_data = {
        "class_number": class_number,
        "subject_id": subject_id,
        "faculty_id": "",  
        "student_data": [],
    }
    for i, student in enumerate(student_data):
        student_marks = marks_data[i]
        student_entry = {"register_number": student}
        if subject_type == "relative":
            student_entry.update({
                "cat1": student_marks["cat1"],
                "cat2": student_marks["cat2"],
                "quiz1": student_marks["quiz1"],
                "quiz2": student_marks["quiz2"],
                "da": student_marks["da"],
                "fat": student_marks["fat"],
                "total_marks": student_marks["total_marks"],
                "grade": student_marks["grade"]
            })
        else:
            student_entry.update({
                "ia1": student_marks["ia1"],
                "mt": student_marks["mt"],
                "ia2": student_marks["ia2"],
                "fat": student_marks["fat"],
                "total_marks": student_marks["total_marks"],
                "grade": student_marks["grade"]
            })
        class_data["student_data"].append(student_entry)
    return class_data
```

```python
# Distributing students into classes

class_number_counter = {sem: 1 for sem in sem_arr} # To track current class number for each semester

for sem in sem_arr:
    for sub in sub_arr:
        subject = subjects_collection.find_one({"subject_id": sub}, {"_id": 0, "type": 1})
        subject_type = subject["type"]

        register_numbers = fetch_student_data(sem, sub)
        
        if subject_type == "relative":
            marks_data = generate_relative_marks(len(register_numbers))
        else:
            marks_data = generate_absolute_marks(len(register_numbers))
        
        num_students = len(register_numbers)
        num_classes = (num_students // 10) + (1 if num_students % 10 != 0 else 0)
        
        start_index = 0
        for i in range(num_classes):
            end_index = min(start_index + 10, len(register_numbers))
            class_number = f"{sem}{'{:03d}'.format(class_number_counter[sem])}"
            student_data = register_numbers[start_index:end_index]
            formatted_class_data = format_class_data(class_number, sub, student_data, marks_data[start_index:end_index], subject_type)
            
            # Insert formatted class data into the classes collection
            classes_collection.insert_one(formatted_class_data)
            
            start_index = end_index
            class_number_counter[sem] += 1

print("Data has been successfully inserted into the classes collection.")
```

```python
def calculate_relative_total_marks(student):
    return (
        0.30 * student["cat1"] +
        0.30 * student["cat2"] +
        student["quiz1"] +
        student["quiz2"] +
        student["da"] +
        0.40 * student["fat"]
    )

def calculate_absolute_total_marks(student):
    return student["ia1"] + student["mt"] + student["ia2"] + student["fat"]
```

```python
def calculate_grade_points(total_marks, mean, sd):
    if total_marks >= mean + 1.5 * sd:
        return 10
    elif (mean + 0.5 * sd) <= total_marks < (mean + 1.5 * sd):
        return 9
    elif (mean - 0.5 * sd) <= total_marks < (mean + 0.5 * sd):
        return 8
    elif (mean - 1.0 * sd) <= total_marks < (mean - 0.5 * sd):
        return 7
    elif (mean - 1.5 * sd) <= total_marks < (mean - 1.0 * sd):
        return 6
    elif (mean - 2.0 * sd) <= total_marks < (mean - 1.5 * sd):
        return 5
    elif total_marks <= mean - 2 * sd:
        return 0
    else:
        return 0
```

```python
# Calculate grade points for each student

classes = classes_collection.find()

for class_doc in classes:
    student_data = class_doc['student_data']
    
    total_marks_list = [student['total_marks'] for student in student_data]
    
    mean_marks = np.mean(total_marks_list)
    sd_marks = np.std(total_marks_list)
    
    for student in student_data:
        total_marks = student['total_marks']
        grade_points = calculate_grade_points(total_marks, mean_marks, sd_marks)
        student['grade'] = grade_points
    
    classes_collection.update_one(
        {'_id': class_doc['_id']},
        {'$set': {'student_data': student_data}}
    )

print("Updated Grade points for all students in all classes.")
```

```python
# Time slots for the timetable
time_slots = [
    "08:00-09:00", "09:00-10:00", "10:00-11:00", "11:00-12:00", "12:00-13:00"
]

weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

def generate_semester_timetable(semester_prefix):
    query = {"class_number": {"$regex": f"^{semester_prefix}"}}
    classes = list(classes_collection.find(query))
    
    timetable = {day: {slot: None for slot in time_slots} for day in weekdays}
    class_index = 0
    
    for class_doc in classes:
        allocated = False
        for day in weekdays:
            if allocated:
                break
            for slot in time_slots:
                if timetable[day][slot] is None:
                    timetable[day][slot] = class_doc["class_number"]
                    allocated = True
                    break

    return timetable

def generate_all_semester_timetables():
    sem_arr = ['F2122', 'W2122', 'F2223', 'W2223', 'F2324', 'W2324']
    all_timetables = {}
    
    for sem in sem_arr:
        timetable = generate_semester_timetable(sem)
        all_timetables[sem] = timetable
    
    return all_timetables
```

```python
def generate_faculty_timetable(faculty_id):
    faculty = faculty_collection.find_one({"faculty_id": faculty_id})
    
    if not faculty:
        print("Faculty not found")
        return
    
    all_semester_timetables = generate_all_semester_timetables()
    faculty_timetable = {}

    for semester_data in faculty["class_data"]:
        semester_name = semester_data["semester_name"]
        class_list = semester_data["class_list"]
        faculty_timetable[semester_name] = {day: {slot: None for slot in time_slots} for day in weekdays}

        semester_timetable = all_semester_timetables.get(semester_name)
        
        if semester_timetable:
            for day, slots in semester_timetable.items():
                for slot, class_number in slots.items():
                    if class_number in class_list:
                        faculty_timetable[semester_name][day][slot] = class_number

    return faculty_timetable
```

```python
def view_update_class_data(faculty_id):
    faculty = faculty_collection.find_one({"faculty_id": faculty_id})
    
    if not faculty:
        print("Faculty not found")
        return
    
    print("Select Semester:")
    for i, semester_data in enumerate(faculty["class_data"], 1):
        print(f"{i}: {semester_data['semester_name']}")
    
    semester_choice = int(input("Enter your choice: ")) - 1
    selected_semester = faculty["class_data"][semester_choice]["semester_name"]
    class_list = faculty["class_data"][semester_choice]["class_list"]

    print(f"\nClasses in Semester {selected_semester}:")
    for class_number in class_list:
        print(class_number)
    
    class_number = input("Enter class number to update: ")
    if class_number not in class_list:
        print("Invalid class number. You can only update your assigned classes.")
        return

    reg_no = input("Enter student register number: ")
    subfield = input("Enter marks field to update: ")
    new_marks = int(input(f"Enter new value for {subfield}: "))
    db['classes'].update_one(
        {"class_number": class_number, "student_data.register_number": reg_no},
        {"$set": {f"student_data.$[elem].{subfield}": new_marks}},
        array_filters=[{"elem.register_number": reg_no}]
    )
    print(f"Updated {subfield} for register number {reg_no} in class {class_number}.")
```

```python
def student_dashboard(register_number):
    while True:
        print("Student Dashboard")
        print("1: Generate Timetable")
        print("2: View Your Classes")
        print("3: Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            generate_student_timetable(register_number)
        elif choice == '2':
            view_student_data(register_number)
        elif choice == '3':
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
def generate_student_timetable(register_number):
    student = students_collection.find_one({"register_number": register_number})
    
    if not student:
        print(f"Student with register number {register_number} not found.")
        return
    
    all_semester_timetables = generate_all_semester_timetables()
    for semester_data in student["semester_data"]:
        semester_name = semester_data["semester_name"]
        subject_list = semester_data["subject_list"]
        
        print(f"Semester: {semester_name}")
        
        if not subject_list:
            print("No subjects registered for this semester.")
            continue
        
        student_timetable = {day: {slot: None for slot in time_slots} for day in weekdays}
        
        for subject_id in subject_list:
            query = {
                "class_number": {"$regex": f"^{semester_name}"},
                "subject_id": subject_id,
                "student_data.register_number": register_number
            }
            classes = list(classes_collection.find(query))
            
            for class_doc in classes:
                class_number = class_doc["class_number"]
                allocated_time = False
                
                for day in weekdays:
                    if allocated_time:
                        break
                    for slot in time_slots:
                        if not student_timetable[day][slot] and all_semester_timetables[semester_name][day][slot] == class_number:
                            student_timetable[day][slot] = class_number
                            allocated_time = True
                            break
        
        for day in weekdays:
            print(f"Day: {day}")
            if all(value is None for value in student_timetable[day].values()):
                print("--------------------")
            else:
                for slot, class_number in student_timetable[day].items():
                    if class_number:
                        print(f"   Time: {slot} - Class Number: {class_number}")
                print("--------------------")
```

```python
def view_student_data(register_number):
    
    student = db.students.find_one({"register_number": register_number})
    
    if student:
        print("Student Information:")
        print(f"Name: {student['name']}")
        print(f"Register Number: {student['register_number']}")
        print("Semester Data:")
        for semester in student.get('semester_data', []):
            print(f"  Semester: {semester['semester_name']}")
            print(f"  Subjects: {', '.join(semester['subject_list'])}")
            print(f"  GPA: {semester['gpa']}")
        print(f"Completed: {', '.join(student.get('completed', []))}")
        print(f"Arrear: {', '.join(student.get('arrear', []))}")
        print(f"CGPA: {student.get('cgpa', 'N/A')}")
        
        view_marks = input("Do you want to view your mark data? (yes/no): ").strip().lower()
        
        if view_marks == 'yes':
            print("Mark Data:")
            for semester in student.get('semester_data', []):
                for subject_id in semester.get('subject_list', []):
                    class_document = db.classes.find_one({
                        "subject_id": subject_id,
                        "student_data.register_number": register_number
                    })
                    
                    if class_document:
                        student_data = next((data for data in class_document['student_data'] if data['register_number'] == register_number), None)
                        
                        if student_data:
                            subject_info = db.subjects.find_one({"subject_id": subject_id})
                            
                            if subject_info:
                                if subject_info['type'] == 'absolute':
                                    ia1 = student_data.get('ia1', '-')
                                    mt = student_data.get('mt', '-')
                                    ia2 = student_data.get('ia2', '-')
                                    fat = student_data.get('fat', '-')
                                    
                                    print(f"  Class: {class_document['class_number']}")
                                    print(f"  Subject: {class_document['subject_id']}")
                                    print(f"    IA1: {ia1}")
                                    print(f"    MT: {mt}")
                                    print(f"    IA2: {ia2}")
                                    print(f"    FAT: {fat}")
                                elif subject_info['type'] == 'relative':
                                    cat1 = student_data.get('cat1', '-')
                                    cat2 = student_data.get('cat2', '-')
                                    quiz1 = student_data.get('quiz1', '-')
                                    quiz2 = student_data.get('quiz2', '-')
                                    da = student_data.get('da', '-')
                                    fat = student_data.get('fat', '-')
                                    
                                    print(f"  Class: {class_document['class_number']}")
                                    print(f"  Subject: {class_document['subject_id']}")
                                    print(f"    CAT1: {cat1}")
                                    print(f"    CAT2: {cat2}")
                                    print(f"    Quiz1: {quiz1}")
                                    print(f"    Quiz2: {quiz2}")
                                    print(f"    DA: {da}")
                                    print(f"    FAT: {fat}")
                                    
                            else:
                                print(f"Subject information not found for {subject_id}")
                    else:
                        print(f"Class not found for subject {subject_id}")
    else:
        print("Student not found.")
```
